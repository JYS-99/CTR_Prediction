```python
def timestamp_change(df):
    df['timestamp_s'] = df['timestamp'].apply(lambda x:int(round(x/1000,0)))
    df['time_processed'] = df['timestamp_s'].apply(lambda x:time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(x)))
    df['year'] = df['time_processed'].apply(lambda x:x[0:4])
    df['month'] = df['time_processed'].apply(lambda x:x[5:7])
    df['day'] = df['time_processed'].apply(lambda x:x[8:10])
    df['hour'] = df['time_processed'].apply(lambda x:x[11:13])
    df['miniute'] = df['time_processed'].apply(lambda x:x[14:16])
    df['second'] = df['time_processed'].apply(lambda x:x[17:19])
    df = pd.concat([df['time_processed'],df['year'],df['month'],df['day'],df['hour'],df['miniute'],df['second']],axis=1)
    return df
```

```python
test_timestamp_notnull = test[test['timestamp'].notnull()]
timestamp_change_df = timestamp_change(test_timestamp_notnull)
timestamp_change_df
```

# 10.Normalization

```python
normalization_lst = ['distance', 'item_ave_price','price', 'user_home_dis', 'user_work_dis','user_displayed_item_num','online_days',
                    'col_4','delta_price']
normalization_df = test_day_hour[normalization_lst]
```

```python
normalization_concat_list = []
for attribute in normalization_lst:
    min_max_scaler = preprocessing.MinMaxScaler()
    x = normalization_df[attribute].values.reshape(-1,1)
    x_minmax = min_max_scaler.fit_transform(x)
    d2 = pd.DataFrame(x_minmax)
    normalization_concat_list.append(d2)
normalization_concat_df = pd.concat(normalization_concat_list,axis=1)
normalization_concat_df.columns = ['distance', 'item_ave_price','price', 'user_home_dis', 'user_work_dis','user_displayed_item_num',
                                   'online_days', 'col_4','delta_price']  #9个
```

# 11.onehot

```python
onehot_lst = ['day', 'hour','weekday','level', 'col_1', 'col_3','col_2', 'gender', 'mobile_os', 'age',
              'dtype','cate_1', 'geo_cat']
onehot_df = test_day_hour[onehot_lst]
```

```python
tail_merged_df
```

```python
def onehot_attributes(dataframe1,attribute):
    df_onehot_output = pd.get_dummies(dataframe1[attribute],prefix=attribute)
    return df_onehot_output

onehot_concat_list1 = []
for attribute in tqdm.tqdm(onehot_lst):
    df = onehot_attributes(onehot_df,attribute)
    onehot_concat_list1.append(df)
    
onehot_concat_df1 = pd.concat(onehot_concat_list1,axis=1)
```

```python
tail_merged_columns_attributes = ['cityid', 'loc_cityid', 'mobile_type', 'cate_2', 'cate_3']

onehot_concat_list2 = []
for attribute in tail_merged_columns_attributes:
    df = onehot_attributes(tail_merged_df,attribute)
    onehot_concat_list2.append(df)
    
onehot_concat_df2 = pd.concat(onehot_concat_list2,axis=1)
```
